# Product Analysis

```python
product_counts = Retail_Sales["Product_Name"].value_counts() # count of each product
product_counts
```

```python
# Top-Selling product by quantity
Top_Selling_product = Retail_Sales.groupby("Product_Name")["Quantity"].sum().sort_values(ascending=False)
Top_Selling_product
```

```python
# top Grossing product by total sales
Top_Grossing_product = Retail_Sales.groupby("Product_Name")["Total_Sales"].sum().sort_values(ascending=False)
Top_Grossing_product
```

```python
# Least sold or least profitable products
Least_sold_product = Retail_Sales.groupby("Product_Name")["Quantity"].sum().sort_values(ascending=True)
Least_sold_product
```

# Category Level Analysis

```python
Category_name = Retail_Sales ["Category"].value_counts() # Count of each Category
Category_name
```

```python
# Total sales per category
Total_sales_per_category = Retail_Sales.groupby("Category")["Total_Sales"].sum()
Total_sales_per_category
```

```python
# Most popular category (by quantity or sales)
Most_popular_category = Retail_Sales.groupby("Category")["Quantity"].sum().sort_values(ascending=False)
Most_popular_category
```

```python
# Average sales per product in each category.
Average_sales_per_product = Retail_Sales.groupby("Category")["Total_Sales"].mean()
Average_sales_per_product
```

# Customer Behavior Analysis

```python
# Customer Behavior Analysis
unique_customers = Retail_Sales['Customer_ID'].nunique()
unique_customers
```

```python
# customers by total spending
top_customers_spending = Retail_Sales.groupby('Customer_ID')['Total_Sales'].sum().sort_values(ascending=False)
top_customers_spending
```

# 5) Time-Based Analysis

```python
from os import RTLD_LAZY
Retail_Sales['Date'] = pd.to_datetime(Retail_Sales['Date'],format='%d-%m-%Y')
Retail_Sales['Month'] = Retail_Sales['Date'].dt.to_period('M')
Retail_Sales['Week'] = Retail_Sales['Date'].dt.isocalendar().week
Retail_Sales['Day'] = Retail_Sales['Date'].dt.date
Retail_Sales['Weekday'] = Retail_Sales['Date'].dt.day_name()
```

```python
Retail_Sales['Year'] = Retail_Sales['Date'].dt.year
Year_sales = Retail_Sales.groupby('Year')['Total_Sales'].sum()
Year_sales
```

```python
monthly_sales = Retail_Sales.groupby('Month')['Total_Sales'].sum()
monthly_sales
```

```python
monthly_sales = Retail_Sales.groupby(Retail_Sales['Date'].dt.to_period('M'))['Total_Sales'].sum()
monthly_sales
```

```python
# Sales by Weekday
weekly_sales = Retail_Sales.groupby('Weekday')['Total_Sales'].sum().sort_values(ascending=False)
weekly_sales
```

#  Visualizations in Graph

```python
# Name of product
product_counts = Retail_Sales['Product_Name'].value_counts()
plt.figure(figsize=(10,10))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%')
plt.title('Product name')
plt.axis('equal')
plt.show()
```

```python
# Bar charts of top products/categories
plt.figure(figsize=(10, 6))
Colour =('DarkBlue', 'Blue', 'MediumBlue', 'DodgerBlue', 'SteelBlue', 'CadetBlue', 'CornflowerBlue', 'SkyBlue', 'LightBlue', 'LightCyan')
# Colour= ('red','blue','green','yellow','orange','purple', 'brown', 'pink', 'gray', 'olive')
plt.bar(Top_Selling_product.index[:10], Top_Selling_product.values[:10], color=Colour)
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.title('Top 10 Selling Products')
plt.xticks(rotation=90)
plt.show()
```

```python
# Line plot of sales over time

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index.astype(str), monthly_sales.values, marker='o', linestyle='-', color='b')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=90) # Rotate labels for better readability
plt.show()
```

```python
# Pie chart of category share

plt.figure(figsize=(10, 6))
plt.pie(Category_name.values, labels=Category_name.index, autopct='%1.1f%%', startangle=90, colors=Colour)
plt.title('Category Share')
plt.show()
```

```python
# total sale graphy
plt.figure(figsize=(10, 6))
plt.bar(Year_sales.index, Year_sales.values, color=Colour)
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Year')
plt.show()
```

```python
# total sales by category graphy
plt.figure(figsize=(10, 6))
Colour = ['red', 'darkred', 'indianred', 'lightcoral', 'salmon'] # Using valid color names
plt.bar(Total_sales_per_category.index, Total_sales_per_category.values, color=Colour)
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Category')
plt.show()
```

```python
# customers by total spending graph
plt.figure(figsize=(10, 6))
plt.plot(top_customers_spending.values[:10], top_customers_spending.index[:10], color='blue',marker='*') # Using a single valid color
plt.xlabel('Customer ID')
plt.ylabel('Total Sales')
plt.title('Top 10 Customers by Total Spending')
plt.show()
```

```python
# Category-wise Monthly Trend
monthly_cat = Retail_Sales.groupby([Retail_Sales['Date'].dt.to_period('M'), 'Category'])['Total_Sales'].sum().unstack()
monthly_cat.plot(figsize=(12,6), title="Monthly Sales Trend by Category")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Heatmap: Weekday vs Category Sales
heatmap_data = Retail_Sales.pivot_table(index='Weekday', columns='Category', values='Total_Sales', aggfunc='sum')
sns.heatmap(heatmap_data, annot=True)
plt.title("Heatmap of Sales by Weekday and Category")
plt.show()
```

```python
#  Boxplot of Price per Category
sns.boxplot(x='Category', y='Price', data=Retail_Sales)
plt.title("Boxplot of Price by Category")
plt.show()
```

```python
# Histogram of Quantity Ordered
Retail_Sales['Quantity'].plot(kind='hist', bins=10)
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.title('Distribution of Quantity')
plt.show()
```

```python
# Cumulative Sales Over Time
Retail_Sales.sort_values('Date', inplace=True)
Retail_Sales['Cumulative_Sales'] = Retail_Sales['Total_Sales'].cumsum()
Retail_Sales.plot(x='Date', y='Cumulative_Sales', title='Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Sales')
plt.show()
```

# 📌 Conclusion
This exploratory data analysis (EDA) on the retail sales dataset provided several valuable insights into product performance, customer behavior, and temporal sales trends. Here are the key takeaways:


#🛍️ Product and Category Insights
Jeans, Wall Art, and Headphones emerged as the top-selling products by quantity, indicating strong and consistent demand.


In terms of revenue, electronics and sports products contributed significantly, showing that high-value items, though fewer in quantity, drive substantial income.

The Groceries category showed frequent purchases with moderate sales per transaction, reflecting regular customer needs.

#👥 Customer Behavior
The business has a solid customer base with 100 unique customers, some of whom made multiple purchases.

A small group of customers (top 10) contributed to a large portion of the revenue, highlighting potential for customer loyalty programs.

Purchase patterns suggest a need for customer segmentation and personalized offers.

#📅 Time-Based Trends
Sales showed clear seasonal variation, with higher sales in certain months, indicating marketing campaigns or festivals might be boosting sales.

Weekday analysis revealed that sales are relatively consistent, but a few specific days perform better — possibly due to paydays or weekend shopping.

Year-over-year trends suggest business growth, providing optimism for future performance if current strategies are maintained or improved.

#📈 Overall Performance
The dataset reflects a healthy balance of frequently bought essentials (like groceries) and high-ticket items (like electronics).

With strong product performance and repeat customers, the business has a good opportunity to scale through data-driven decision-making, targeted marketing, and stock optimization.
